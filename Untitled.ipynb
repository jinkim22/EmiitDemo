```python
import pandas as pd
from datetime import datetime
from sklearn.model_selection import train_test_split
import numpy as np
dateparse = lambda x: datetime.strptime(x, '%Y-%m')
```

```python
df = pd.read_csv("emission.csv")
df = df.dropna()
```

```python
# df = pd.read_csv("emission.csv",  parse_dates=['Date'], date_parser=dateparse)
```

```python
df = df.set_index('Date')
```

```python
from statsmodels.tsa.stattools import adfuller

df.columns = df.columns.str.replace(' ','_')
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, df['three_month_prediction'], test_size = 0.25)
```

```python
X_train.head()
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000)
# Train the model on training data
rf.fit(X_train, y_train);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(X_test)
# Calculate the absolute errors
errors = (abs(predictions - y_test) / y_test)
# Print out the mean absolute error (mae)
```

```python
sum(errors) / len(errors)
```

```python
# Get numerical feature importances
importances = list(rf.feature_importances_)
# List of tuples with variable and importance
feature_importances = [(feature, importance) for feature, importance in zip(df.columns[:-1], importances)]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];
```
